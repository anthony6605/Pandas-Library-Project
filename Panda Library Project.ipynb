```python
age_groups = ['Age 16-19', 'Age 20-24', 'Age 25-34', 'Age 35-44','Age 45-54','Age 55-64','Age 65+']
```

```python
for ages in age_groups:
    plt.figure(figsize=(6, 3))
    plt.hist(df[ages], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {ages}')
    plt.xlabel('Age Group')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

```python
for ages in age_groups:
    plt.figure(figsize=(10, 6))
    df.boxplot(column=ages, by='Year')
    plt.title(f'{ages} by Year')
    plt.xlabel('Year')
    plt.ylabel('Age Group')
    plt.xticks(rotation = 45)
    plt.grid(True)
    plt.show()
```
